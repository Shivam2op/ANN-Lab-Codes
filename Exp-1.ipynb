**Exp 1: Plot various types of activation functions used in neural network**

```python
import matplotlib.pyplot as plt
import numpy as np

def linear(x):
    return x

x = np.linspace(-10, 10, 100)
y = linear(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Linear Activation Function")
plt.grid()
plt.show()
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Sigmoid Activation Function")
plt.grid()
plt.show()
```

```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-10 , 10, 100)
y = relu(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Relu Activation Function")
plt.grid()
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 100)
y = tanh(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Tanh Activation Function")
plt.grid()
plt.show()
```

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis= 0)

x = np.linspace(-10, 10, 100)
y = softmax(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Softmax Activation Function")
plt.grid()
plt.show()
```

```python
def leaky_relu(x, alpha= 0.02):
    return np.maximum(alpha * x,x)

x = np.linspace(-10, 10, 100)
y = leaky_relu(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Leaky Relu Activation Function")
plt.grid()
plt.show()
```
